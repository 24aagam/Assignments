# 3. Explain the following terms:

**Ans:**
- **The difference in np.array, np.asarray and np.asanyarray.**
 - Use `np.array()` when creating an array from scratch and you need more options.
 - Use `np.asarray()` when you want to work with the input data as an array without unnecessary copying.
 - Use `np.asanyarray()` when you want to preserve the subclass of ndarray (e.g., an np.matrix), it returns the same subclass or behave like np.`asarray()` otherwise

- **The difference between Deep copy and shallow copy.**
 - **Shallow Copy:**
   - A shallow copy of an object creates a new object, but it shares the same memory with the original object for its elements.
   - It copies the structure, not the actual data.
   - Changes made to the elements in the shallow copy are reflected in the original object as well.
   - Shallow copy is faster because it doesn’t duplicate the data.
   - **Example:** If you create a shallow copy of a list, both the original list and the copy will point to the same list elements.
 - Deep Copy:
   - A deep copy creates a completely independent copy of the original object.
   - It clones the entire data structure, including nested objects.
   - Changes made to the deep copy do not affect the original object.
   - Deep copy is slower because it recursively duplicates all data.
   - **Example:** If you create a deep copy of a list, it creates a new list with its own separate elements.

# 17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

```python
arr1 = np.random.rand(100)
arr2 = np.random.rand(100)
```

```python
arr1
```

```python
arr2
```

```python
import seaborn as sns
```

```python
quadrant = np.zeros_like(arr1)
quadrant[(arr1 > 0) & (arr2 > 0)] = 1  # First quadrant
quadrant[(arr1 < 0) & (arr2 > 0)] = 2  # Second quadrant
quadrant[(arr1 < 0) & (arr2 < 0)] = 3  # Third quadrant
quadrant[(arr1 > 0) & (arr2 < 0)] = 4  # Fourth quadrant
```

```python
sns.scatterplot(x=arr1, y=arr2, hue=quadrant, palette='Set1',legend='full' )
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Quadrant-wise Scatter Plot")
```

# 19. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'.

```python
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256

# Generate random categorical data
categories = ["A", "B", "C"]
values = [10, 20, 15]  # You can replace these with your own data

# Create a DataFrame
df = pd.DataFrame({"Category": categories, "Value": values})

# Create a color mapper based on values
mapper = linear_cmap(field_name="Value", palette=Viridis256, low=min(values), high=max(values))

# Create a Bokeh figure
p = figure(x_range=categories, height=400, title="Random Categorical Bar Chart")  # Use 'height' instead of 'plot_height'
source = ColumnDataSource(df)

# Add bars with color mapping
p.vbar(x="Category", top="Value", source=source, width=0.5, color=mapper, legend_label="Values")

# Add hover tooltips
hover = HoverTool()
hover.tooltips = [("Category", "@Category"), ("Value", "@Value")]
p.add_tools(hover)

# Label axes
p.xaxis.axis_label = "Categories"
p.yaxis.axis_label = "Values"

# Show the plot
output_notebook()  # For Jupyter Notebook
show(p)
```
